```python
import pandas as pd

df = pd.read_csv('Apple Music Play Activity.csv')
```

```python
pd.set_option('display.max_columns', None)
df.head()
```

```python
filtered_df = df[['Age Bucket', 'Album Name', 'Apple Music Subscription', 'Build Version', 'Client Platform', 'Container Album Name', 'Container Artist Name', 'Container Type', 'End Reason Type', 'Event End Timestamp', 'IP City', 'Milliseconds Since Play', 'Play Duration Milliseconds', 'Repeat Play', 'Shuffle Play', 'Song Name']]

filtered_df.head(3)
```

```python
music = filtered_df.copy()

music.rename(columns = {'Play Duration Milliseconds': 'play_duration'}, inplace = True)

music['play_duration_minutes'] = music['play_duration'] / 1000 / 60

music
```

```python
music.groupby('Song Name')['play_duration_minutes'].sum().sort_values(ascending = False).reset_index()
```

```python
music.groupby('Song Name')['play_duration_minutes'].agg(['sum', 'mean']).sort_values(['sum', 'mean'], ascending = [False, False]).reset_index()
```

```python
music.groupby(['Song Name', 'IP City'])['play_duration_minutes'].agg(['count', 'sum', 'mean']).sort_values(['count', 'sum', 'mean'], ascending = [False, False, False]).reset_index()
```

```python
(music['IP City'].value_counts(normalize = True) * 100).sort_values(ascending = False)
```

```python
music[music['IP City'] == 'LIVINGSTON'].groupby('Song Name')['play_duration_minutes'].agg(['size', 'mean', 'sum']).sort_values(['size'], ascending = [False]).head(10)
```
